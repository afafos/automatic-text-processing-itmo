# Task 8: networkx (плотность, кратчайший путь, максимальная клика)

В этом задании:
- создается случайный плотный неориентированный граф `G = nx.dense_gnm_random_graph(53, 656, seed=44)`;
- вычисляется плотность графа;
- находится длина кратчайшего пути между вершинами `3` и `47`;
- определяется размер максимальной клики.

```python
import networkx as nx
from networkx.algorithms.clique import find_cliques

G = nx.dense_gnm_random_graph(43, 255, seed=86)

density = round(nx.density(G), 3)
shortest_path_len = nx.shortest_path_length(G, source=3, target=30)
max_clique_size = max(len(clique) for clique in find_cliques(G))

print(density)
print(shortest_path_len)
print(max_clique_size)
```
